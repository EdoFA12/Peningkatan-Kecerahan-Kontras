```python
##Melakukan Import library yang di butuhkan
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from google.colab import files
import matplotlib.pyplot as plt
```

```python
##Upload Citra
uploaded = files.upload()
file_name = next(iter(uploaded))

image = cv2.imread(file_name)

##Memberikan Validasi pada Image Apakah image berhasil di baca apa tidak oleh lib openCV
if image is None:
    print("❌ ERROR: Citra gagal dimuat. Pastikan nama file benar.")
else:
    # Tampilkan citra asli
    print(f"✅ Citra Asli: {file_name} ({image.shape})")
    cv2_imshow(image)
```

```python
##Implementasi Peningkatan Kecerahan/Kontras
alpha = 1.5
beta = 30

adjusted_image = cv2.convertScaleAbs(image, alpha=alpha, beta=beta)
print(f"✅ Citra Hasil Olahan (Kontras: {alpha}, Kecerahan: {beta})")
cv2_imshow(adjusted_image)
```
